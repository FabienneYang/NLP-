```python
#運用scikit-learn API 實現K-fold分割資料
import pandas as pd
dataset = pd.read_csv(r'Social_Network_Ads.csv')
```

```python
dataset
```

```python
#取出訓練特徵與標註
X = dataset[['User ID', 'Gender', 'Age', 'EstimatedSalary']].values
Y = dataset['Purchased'].values
```

```python
import numpy as np
from sklearn.model_selection import KFold
```

```python
#將訓練資料按照順序切割成10等分
kf = KFold(n_splits = 10)
kf.get_n_splits(X)

print(kf)
```

```python
#將訓練資料隨機切割成10等分
kf = KFold(n_splits = 10, shuffle=True)
kf.get_n_splits(X)

print(kf)
```

```python
#取出 切割資料對應位置
train_split = kf.split(X)
next(train_split)
```

```python
#取出切割資料：trainset / testset 特徵(x_train/x_test)/標註(y_train/y_test)

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = Y[train_index], Y[test_index]
```
